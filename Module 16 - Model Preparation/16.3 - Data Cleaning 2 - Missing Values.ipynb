### Question 2

If enrollment or revenue data is needed for certain analysis, it's probably best to discard observations that don't have data for them. State budgets developed to local education will vary greatly by size, local politics, and educational institutions in the area. An educated guess will likely be far off. 

For missing test scores, imputing a median value for that test probably makes the most sense.

```python
used_df_2 = used_df.fillna(used_df.median())
```

### Question 3



```python
used_df_3 = pd.DataFrame(columns = used_df.columns)

for y in range(1992,2018):
    used_df_3 = used_df_3.append(used_df[used_df['YEAR']==y].fillna(used_df[used_df['YEAR']==y].median()))
```
