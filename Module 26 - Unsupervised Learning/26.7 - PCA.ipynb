# 26.7 - Dimensionality Reduction with PCA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
import warnings
from sklearn import datasets, metrics
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_openml
```

```python
mnist = fetch_openml('Fashion-MNIST', version=1, cache=True)
```

```python
np.random.seed(23)

indices = np.random.choice(70000,10000)
X = mnist.data[indices]
y = mnist.target[indices]

plt.figure(figsize=(20,12))

plt.subplot(2,4,1)
plt.imshow(X[0].reshape(28,28))
plt.title('Image: {}'.format(y[0]))

plt.subplot(2,4,2)
plt.imshow(X[1].reshape(28,28))
plt.title('Image: {}'.format(y[1]))

plt.subplot(2,4,3)
plt.imshow(X[2].reshape(28,28))
plt.title('Image: {}'.format(y[2]))

plt.subplot(2,4,4)
plt.imshow(X[3].reshape(28,28))
plt.title('Image: {}'.format(y[3]))

plt.subplot(2,4,5)
plt.imshow(X[4].reshape(28,28))
plt.title('Image: {}'.format(y[4]))

plt.subplot(2,4,6)
plt.imshow(X[5].reshape(28,28))
plt.title('Image: {}'.format(y[5]))

plt.subplot(2,4,7)
plt.imshow(X[6].reshape(28,28))
plt.title('Image: {}'.format(y[6]))

plt.subplot(2,4,8)
plt.imshow(X[7].reshape(28,28))
plt.title('Image: {}'.format(y[7]))

plt.show()
```

```python
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

pca = PCA(n_components=2).fit_transform(X_std)

plt.figure(figsize=(12,6))
plt.scatter([x[0] for x in pca], [x[1] for x in pca])
plt.show()
```

```python
colors = ["r","b","g","c","m","y","k","r","burlywood","chartreuse"]

plt.figure(figsize=(12,6))
plt.scatter([x[0] for x in pca], [x[1] for x in pca], c=[colors[int(x)] for x in y])
plt.show()
```

Q: Do you think PCA solution is satisfactory? Can you distinguish between different classes easily?

It depends what you mean by "distinguish." The classes defintely occupy distinct areas, but there is a lot of overlap. Most areas of the chart have a mix of several classes, and it looks like at least one class (red, or t-shirt/top) has two separated areas that it occupies. If you were using this solution to identify class, it appears that you would get things wrong often.
