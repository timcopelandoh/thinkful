```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

rvar = np.random.poisson(3,1000)
plt.hist(rvar, bins = 30)
plt.axvline(rvar.mean(), color = 'y', linewidth = 2)
plt.axvline(rvar.mean()+rvar.std(), color = 'y', linestyle = 'dashed', linewidth = 2)
plt.axvline(rvar.mean()-rvar.std(), color = 'y', linestyle = 'dashed', linewidth = 2)

plt.title('Poisson Distribution')
plt.show()
```

Even though the distribution is non-negative, discrete, and skewed, I think mean and standard deviation do a fairly good job of describing the data.

```python
rvar = np.random.logistic(3,1,1000)
plt.hist(rvar, bins = 30)
plt.axvline(rvar.mean(), color = 'y', linewidth = 2)
plt.axvline(rvar.mean()+rvar.std(), color = 'y', linestyle = 'dashed', linewidth = 2)
plt.axvline(rvar.mean()-rvar.std(), color = 'y', linestyle = 'dashed', linewidth = 2)

plt.title('Logistic Distribution')
plt.show()
```

I think mean and standard deviation do a good job.

```python
rvar = np.random.lognormal(0,1,1000)
plt.hist(rvar, bins = 30)
plt.axvline(rvar.mean(), color = 'y', linewidth = 2)
plt.axvline(rvar.mean()+rvar.std(), color = 'y', linestyle = 'dashed', linewidth = 2)
plt.axvline(rvar.mean()-rvar.std(), color = 'y', linestyle = 'dashed', linewidth = 2)

plt.title('Log Normal Distribution')
plt.show()
```

I don't think that mean and standard deviation do a great job of describing this data. Mean could be fairly useful, but the data is non-negative, and a standard deviation below the mean extends past where there are any observed values. A measure of how fast the data "decays," may be more useful.

```python
rvar = np.random.chisquare(1,1000)
plt.hist(rvar, bins= 30)
plt.axvline(rvar.mean(), color = 'y', linewidth = 2)
plt.axvline(rvar.mean()+rvar.std(), color = 'y', linestyle = 'dashed', linewidth = 2)
plt.axvline(rvar.mean()-rvar.std(), color = 'y', linestyle = 'dashed', linewidth = 2)

plt.title('Chi-Square Distribution')
plt.show()
```

This distribution is very similar to the log-normal distribution. I don't think that mean and standard deviation do a great job of describing this data. A measure of how fast the data "decays," may be more useful.

```python
rvar1 = np.random.normal(5,0.5,1000)
rvar2 = np.random.normal(10,1,1000)
rvar = rvar1+rvar2
plt.hist(rvar, bins = 30)
plt.axvline(rvar.mean(), color = 'y', linewidth = 2)
plt.axvline(rvar.mean()+rvar.std(), color = 'y', linestyle = 'dashed', linewidth = 2)
plt.axvline(rvar.mean()-rvar.std(), color = 'y', linestyle = 'dashed', linewidth = 2)

plt.title('Sum of Two Normal Distributions')
plt.show()
```

I think mean and standard deviation describe the data well.
