### Question 1: Weather model

```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

postgres_user = 'dsbc_student'
postgres_pw = '7*.8G9QH21'
postgres_host = '142.93.121.174'
postgres_port = '5432'
postgres_db = 'weatherinszeged'

engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(
    postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))

weather = pd.read_sql_query('select * from weatherinszeged',con=engine)

engine.dispose()
```

```python
y = weather['apparenttemperature'] - weather['temperature']
X = weather[['humidity', 'windspeed']]

results = sm.OLS(y, X).fit()
results.summary()
```

```python
X = sm.add_constant(X)

results = sm.OLS(y, X).fit()
results.summary()
```

I wasn't sure whether there was supposed to be a constant in the model. My intuition would say that it would make more sense to do so. Without the constant added in, the r-squared value is .425, and both the AIC and BIC are 3.535 * 10^5. With the constant added in, the r-squared decreases to .288. This seems surprising, for the amount of explained variance to decrease after adding in a variable, especially one as intuitive as a constant term. However, the AIC and BIC also decrease, to 3.409 * 10^5. 

Variables such as temperature and wind chill/heat index should be fairly predictable. Even an r-squared of .425 is a little disappointing.

```python
weather['humidity*windspeed'] = weather['humidity']*weather['windspeed']

X = weather[['humidity', 'windspeed', 'humidity*windspeed']]

results = sm.OLS(y,X).fit()
results.summary()
```

```python
X = weather[['humidity', 'windspeed', 'visibility']]

results = sm.OLS(y, X).fit()
results.summary()
```

Using all measures of model performance, adding in the interaction term improves performance.

Adding visibility to the model instead improves the main model, but not to the main extent that the interaction term does.

### Question 2: Housing model

```python
postgres_db = 'houseprices'

engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(
    postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))

house = pd.read_sql_query('select * from houseprices',con=engine)

engine.dispose()

y = house['saleprice']

X = house[['grlivarea', 'totalbsmtsf', 'fullbath', 'halfbath', 'overallqual', 'overallcond', 'yearbuilt', 'garagearea']]
X = sm.add_constant(X)

results = sm.OLS(y, X).fit()

results.summary()
```

```python
X = house[['grlivarea', 'totalbsmtsf', 'fullbath', 'halfbath', 'overallqual', 'overallcond', 'yearbuilt', 'garagearea']]

results = sm.OLS(y, X).fit()

results.summary()
```

This is another weird result with regard to the constant term. After removing the constant, the r-squared increases drastically, to .961, but the AIC and BIC increase, indicating worse performance.

```python
house['paveddrive'] = house['paveddrive'].apply(lambda x: [0, 1][(x=='Y')])
house['centralair'] = house['centralair'].apply(lambda x: [0, 1][(x=='Y')])

X = house[['grlivarea', 'totalbsmtsf', 'fullbath', 'halfbath', 'overallqual', 'overallcond', 'yearbuilt', 'garagearea', 'poolarea', 'paveddrive', 'centralair']]

results = sm.OLS(y, X).fit()

results.summary()
```

```python
X = sm.add_constant(X)

results = sm.OLS(y, X).fit()

results.summary()
```

Adding in pool square footage, whether a house has a paved driveway, and whether a house has central air conditioning does not improve our model. The adjusted r-squared values remain the same, and 2 of the 4 AIC/BIC values get worse, with one improving and one staying the same.
